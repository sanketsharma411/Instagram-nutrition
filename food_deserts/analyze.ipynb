```python
def read_post_line(line):
    line = line.split('\t');
    return {'tract':line[0],'post_id':line[1],'tags':line[2].split(','),'time':int(float(line[3])), 'lat':float(line[4]), 'lon':float(line[5]), 'match_tags':line[6].split(','), 'match_ids':line[7].split(','),\
            'cal':float(line[8]), 'sug':float(line[9]),'fat':float(line[10]), 'cho':float(line[11]), 'fib':float(line[12]), 'pro':float(line[13]),\
            'tract_long':line[14], 'fd':int(line[15])}
```

```python
def write_dict_to_file(d,filename,sort_by_val=False):
    '''
    Function writes the passed dictionary to the filename, with one key value pair at each line, unformatted
    Appends to end of file
    '''
    if sort_by_val:
        sorted_dict = sorted(d.items(), key = lambda x:x[1], reverse = True)
        d = {k:v for k,v in sorted_dict}
    with open(filename,'a+') as f:
        st = ''
        for k in d:
            st += str(k)+','+str(d[k])+'\n'
        f.write(st)
    print 'Done Writing to',filename
```

### Canonical food names and their frequency of appearance in non food desert posts and food desert posts.
Generates the following dictionaries
+ fd_can_count Count of Canonical Names
+ fd_id_count   Count of USDA IDs
+ fd_tag_count  Count of tags


```python
fd_can_count = defaultdict(int)
fd_id_count= defaultdict(int)
fd_tag_count= defaultdict(int)
c = 0
for tract in fd_tract_post_dict:
    c = c+ 1
    if not c%1000: print c
    for post in fd_tract_post_dict[tract]:
        for tag in post['tags']:
            tag = p.singular_noun(tag) if p.singular_noun(tag) != False else tag
            fd_tag_count[tag] += 1
            
        for can_name in post['match_tags']:
            fd_can_count[can_name]+= 1
            
        for usda_id in post['match_ids']:
            fd_id_count[usda_id]+= 1
print 'Tags :',len(fd_tag_count),'Canonical Names : ',len(fd_can_count),'IDS : ',len(fd_id_count)
# Tags : 106028 Canonical Names :  1234 IDS :  8554
# After singularization
# Tags : 88803 Canonical Names :  1234 IDS :  8554
```

### Difference : Canonical Names
##### Printing the difference/sum in the percentages of fd and non-fd for the words in our canonical list
The counts are normalized by the number of samples
i.e. (a-b)/(a+b)

```python
# Loading the Canonical Names file
with open('C:/Users/Shweta/Desktop/CS 8903/FINAL/food_list/canonical_names_list_updated_lower_singular.txt','r') as f:
    lines = f.readlines()
can_list = [line.strip() for line in lines]
```

```python
can_diff_list = {}
for can in can_list:
    if can in fd_can_count and can in non_fd_can_count:
        # Threshold put at 100 and 200 for fd and no-fd respectively
        # so it only looks at canonical matches which appear in at least 100 posts in either of the lists
        if fd_can_count[can]>100 or non_fd_can_count[can]>300:
            fd_percent = fd_can_count[can]*100.0/num_fd_posts
            non_fd_percent = non_fd_can_count[can]*100.0/num_non_fd_posts
            can_diff_list[can] = (fd_percent-non_fd_percent)/(fd_percent+non_fd_percent)
```
